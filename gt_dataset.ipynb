## Imports

```python
import os
import io
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## GT Dataset

```python
path = "highway/input/"
output_path = "highway/input/" 


kernel = np.ones((5,5),np.uint8)

imatges = os.listdir(path)
imatges.sort()

imatge_p = cv2.imread(path + "/" + imatges[0])
imatge_p = imatge_p.astype(np.int16)


for i in range(1, 470): # TODO: read from ROI file
    
    imatge = cv2.imread(path + "/" + imatges[i])
    imatge = imatge.astype(np.int16)

    dif = np.abs(imatge - imatge_p)
    dif = dif.astype(np.uint8)
    dif = cv2.cvtColor(dif, cv2.COLOR_BGR2GRAY)
    
    
    dif = cv2.dilate(dif, kernel, iterations = 1)
    dif = cv2.erode(dif, kernel, iterations = 2)
    
    _, thresh = cv2.threshold(dif, 30, 255, 0)
    contours, _ = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    for c in contours:
        rect = cv2.boundingRect(c)
        cv2.rectangle(thresh, (rect[0],rect[1]),(rect[0]+rect[2],rect[1]+rect[3]),color=255, thickness=-1)
    
    imatge_p = np.copy(imatge)
        
    cv2.imwrite(output_path + "gt" +imatges[i][2:-3]+"png", thresh)
```
